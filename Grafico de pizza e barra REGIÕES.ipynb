```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

colunasEnem=['SG_UF_RESIDENCIA','NU_NOTA_CH','NU_NOTA_LC','NU_NOTA_MT', 'NU_NOTA_CN','NU_NOTA_REDACAO']
enem = pd.read_csv('MICRODADOS_ENEM_2018.csv', sep=";", encoding='ISO-8859-1', usecols= colunasEnem)

cnMedia = enem['NU_NOTA_CN'].mean()
cnMedia = round(cnMedia, 2)
enem.update(enem['NU_NOTA_CN'].fillna(cnMedia))

chMedia = enem['NU_NOTA_CH'].mean()
chMedia = round(chMedia, 2)
enem.update(enem['NU_NOTA_CH'].fillna(chMedia))

lcMedia = enem['NU_NOTA_LC'].mean()
lcMedia = round(lcMedia, 2)
enem.update(enem['NU_NOTA_LC'].fillna(lcMedia))

mtMedia = enem['NU_NOTA_MT'].mean()
mtMedia = round(mtMedia, 2)
enem.update(enem['NU_NOTA_MT'].fillna(mtMedia))

redMedia = enem['NU_NOTA_REDACAO'].mean()
redMedia = round(redMedia, 2)
enem.update(enem['NU_NOTA_REDACAO'].fillna(redMedia))


def regiao(num):
    
    if num in ['SP','RJ','ES','MG']:
        return 'SUDESTE'
    if num in ['PB','PE','RN','MA','CE','BA','SE','PI']:
        return 'NORDESTE'
    if num in ['SC','PR','RS']:
        return 'SUL'
    if num in ['GO','MT','MS']:
        return 'CENTRO OESTE'
    if num in ['TO', 'RR','PA','AC','RO','AM','AP']:
        return 'NORTE'
    
enem['SG_UF_RESIDENCIA'] = enem['SG_UF_RESIDENCIA'].map(regiao)

sudeste = enem.loc[enem['SG_UF_RESIDENCIA'] == 'SUDESTE']
nordeste = enem.loc[enem['SG_UF_RESIDENCIA'] == 'NORDESTE']
sul = enem.loc[enem['SG_UF_RESIDENCIA'] == 'SUL']
norte = enem.loc[enem['SG_UF_RESIDENCIA'] == 'NORTE']
centroOeste=enem.loc[enem['SG_UF_RESIDENCIA'] == 'CENTRO OESTE']

sudesteMedia=[]
nordesteMedia=[]
sulMedia=[]
norteMedia=[]
centrooesteMedia=[]

for i in range(6):
    if i == 0:
        pass
    else:
        sudesteMedia.append(sudeste[colunasEnem[i]].mean())
        nordesteMedia.append(nordeste[colunasEnem[i]].mean())
        sulMedia.append(sul[colunasEnem[i]].mean())
        norteMedia.append(norte[colunasEnem[i]].mean())
        centrooesteMedia.append(centroOeste[colunasEnem[i]].mean())
        
labels=['Ciências Humanas', 'Linguagens e Código', 'Matemática','Ciências da Natureza', 'Redação']

def func(pct, value): 
    porcentagem = float(pct/100*np.sum(value))
    return "{:.2f}%\n({:.2f})".format(pct, porcentagem)

c = ['#65ADB2', '#A37E9D', '#E0D292', '#C9DD8D','#E59C9C']
```

```python
explode = (0.1,0,0,0,0)

fig1, ax1 = plt.subplots(figsize=(6,6))

ax1.pie(sudesteMedia, explode = explode, labels=labels, autopct = lambda pct: func(pct,sudesteMedia), shadow= True, startangle=90, colors=c)
ax1.axis('equal')
ax1.set_title("Média das notas do Enem na região Sudeste")
plt.show()
        
```

```python
explode = (0.1,0,0,0,0)

fig1, ax1 = plt.subplots(figsize=(6,6))

ax1.pie(nordesteMedia, explode = explode, labels=labels, autopct = lambda pct: func(pct,nordesteMedia), shadow= True, startangle=90, colors=c)
ax1.axis('equal')
ax1.set_title("Média das notas do Enem na região Nordeste")
plt.show()
```

```python
explode = (0.1,0,0,0,0)

fig1, ax1 = plt.subplots(figsize=(6,6))

ax1.pie(sulMedia, explode = explode, labels=labels, autopct = lambda pct: func(pct,sulMedia), shadow= True, startangle=90, colors=c)
ax1.axis('equal')
ax1.set_title("Média das notas do Enem na região Sul")
plt.show()
```

```python
explode = (0.1,0,0,0,0)

fig1, ax1 = plt.subplots(figsize=(6,6))

ax1.pie(centrooesteMedia, explode = explode, labels=labels, autopct = lambda pct: func(pct,centrooesteMedia), shadow= True, startangle=90, colors=c)
ax1.axis('equal')
ax1.set_title("Média das notas do Enem na região Centro-Oeste")
plt.show()
```
