```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

colunas_enem = ['NU_NOTA_LC','Q006']
    
quest_nota = pd.read_csv('MICRODADOS_ENEM_2018.csv',sep=';',encoding = 'ISO-8859-1',usecols = colunas_enem)

quest_nota.rename({'Q006':'RENDA FAMILIAR', 'NU_NOTA_LC':'NOTA LINGUAGENS'},inplace = True, axis='columns')

sns.relplot(x="RENDA FAMILIAR", y="NOTA LINGUAGENS", data = quest_nota,kind="line") #grafico relacionando as notas com a renda
plt.show()
```
